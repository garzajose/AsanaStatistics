# Completion by project / by Assignee


```python
uptotoday['completed_at'] = pd.to_datetime(uptotoday['completed_at']).apply(lambda x: x.strftime('%Y-%m-%d')if not pd.isnull(x) else '')
```

```python
uptotoday['completed_at'] = pd.to_datetime(uptotoday['completed_at'])
```

```python
grouped_assignee = uptotoday.groupby('Assignee')[["Task ID", "completed_at"]].count()
```

```python
grouped_completed = uptotoday.groupby('Projects')[["Task ID", "completed_at"]].count()
grouped_completed['Pending']=grouped_completed['Task ID']-grouped_completed['completed_at']
```

```python
del(grouped_completed['Task ID'])
```

```python
grouped_completed.plot(kind='barh', stacked=True, figsize=[16,6], colormap='winter')
```

```python
grouped_assignee['Pending']=grouped_assignee['Task ID']-grouped_assignee['completed_at']
del(grouped_assignee ['Task ID'])
```

```python
grouped_assignee.plot(kind='barh', stacked=True, figsize=[16,6], colormap='winter')
```
